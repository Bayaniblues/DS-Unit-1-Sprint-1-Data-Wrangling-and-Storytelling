```python
import pandas as pd
import numpy as np
import os

csv_list =[]
names_list =[]

split_me = lambda string, split_item : string.split(split_item, 1)

def remove_ext(X):
  split_me(X, ".")
  substring = split_me(X, ".")[0]
  return substring

#remove_ext('test.csv')

directory = '/content/instacart_2017_05_01'
# Too lazy to write everything, using loop
for filename in os.listdir(directory):
    if filename.endswith(".csv"):
      if filename.startswith('._'):
        pass
      else:
        csv_list.append(filename)
        names_list.append(remove_ext(filename))
        continue
    else:
      pass

print(csv_list)
print(names_list)
```

```python
df = []

for x in csv_list:
  for y in names_list:
    if remove_ext(x) == y:
      df.append(pd.read_csv('/content/instacart_2017_05_01/' + x))
    else:
      pass

print(df)
```
